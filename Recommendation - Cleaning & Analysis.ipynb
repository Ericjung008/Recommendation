```python
def comparision(data,col1,col2):
    n = 0
    copy_data = data.copy()
    copy_data.dropna(inplace = True)
    copy_data.reset_index(drop = True, inplace = True)
    
    for i in range(len(copy_data)):
        if copy_data.loc[i,col1] != copy_data.loc[i,col2]:
            n +=1
        
    print('There are {} values that do not match.'.format(n))
```

```python
comparision(products,'Order_Items.product_id','Products.id')
```

```python
def mismatch_df(data,col1,col2):
    list_of_indices = []
    for i in range(len(data)):
        if data.loc[i,col1] != data.loc[i,col2]:
            list_of_indices.append(i)
        
    return data.iloc[list_of_indices]  
```

**Let's check if columns Order_Items.cost and Products.cost are similar.**

```python
def find_error(data, col1, col2):
    reduced_df = data[[col1, col2]]
    reduced_df.dropna(axis = 0, inplace = True)
    
    summation = sum([(x - y) ** 2 for x,y in zip(reduced_df[col1], reduced_df[col2])])
    RMSE = np.sqrt(summation / len(reduced_df))
    
    return RMSE
```

```python
find_error(rec,'Order_Items.cost','Products.cost')
```

**Number of times a product name appears where id is null.**

```python
def counter_df(data,col_id,col_name):
    list_of_names = list(data.loc[data[col_id].isnull(),col_name].values)
    counter_list = [len(data[data[col_name] == name]) for name in list_of_names]
        
    df_counter = pd.DataFrame({'product name': list_of_names, 'number of occurrences': counter_list})
    df_counter.sort_values(by = 'number of occurrences', ascending = False, inplace = True)
    
    return df_counter
```

**Pull the columns needed to find the most popular product.**

```python
popular = rec[['product id','product name', 'quantity', 'price', 'cost']]
popular.head()
```

**Most popular product is the one that brings in the highest profit.**

```python
popular['profit_per_qty'] = [(price - cost) for price,cost in zip(popular['price'],popular['cost'])]
popular['profit'] = popular['profit_per_qty'] * popular['quantity']
```

```python
popular_rec = popular.groupby('product id').agg({'product name': 'first', 'profit': 'sum'})
popular_rec.sort_values(by = 'profit',ascending = False, inplace = True)
popular_rec.rename(columns = {'profit': 'total profit'}, inplace = True)
popular_rec.head(5)
```
